## Drawing line plot: 

```python
import seaborn as sns
import matplotlib.pyplot as plt
#load dataset
flower = sns.load_dataset("iris")
#draw a line plot:
sns.lineplot(x ="petal_width",y = "petal_length",data = flower)
plt.show()
```

### Adding titles:

```python
import seaborn as sns
import matplotlib.pyplot as plt
#load dataset:
flower = sns.load_dataset("iris")
#draw a line plot:
sns.lineplot(x ="petal_width",y = "petal_length",data = flower)
#Adding titles:
plt.title("Graph of flower petal")
plt.show()
```

### Adding limits:

```python
import seaborn as sns
import matplotlib.pyplot as plt
#load dataset
flower = sns.load_dataset("iris")
#draw a line plot:
sns.lineplot(x ="sepal_length",y = "sepal_width",data = flower)
#Adding titles:
plt.title("Graph of flower petal")
#adding limits:
plt.xlim(2.5)
plt.ylim(3)
plt.show()
```

### Set styles:

- darkgrid 
- whitegrid 
- dark
- white
- ticks 

```python
import seaborn as sns
import matplotlib.pyplot as plt
#load dataset
flower = sns.load_dataset("iris")
#changing style:
sns.set_style("darkgrid")
#draw a line plot:
sns.lineplot(x ="sepal_length",y = "sepal_width",data = flower)
#Adding titles:
plt.title("Graph of flower petal")

plt.show() 
```

### Changing size of figure:



```python
import seaborn as sns
import matplotlib.pyplot as plt
#load dataset
flower = sns.load_dataset("iris")
#changing figure size:
plt.figure(figsize= (5,4))
#draw a line plot:
sns.lineplot(x ="petal_width",y = "petal_length",data = flower,color = "red")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
#load dataset
flower = sns.load_dataset("iris")
# style:
sns.set_style(style = "darkgrid")
#changing figure size:
plt.figure(figsize= (5,4))
#draw a line plot:
sns.lineplot(x ="petal_width",y = "petal_length",data = flower,color = "red")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
#load dataset
flower = sns.load_dataset("iris")
sns.set_style(style ="whitegrid")
#changing figure size:
plt.figure(figsize= (5,4))
#draw a line plot:
sns.lineplot(x ="petal_width",y = "species",data = flower,color = "red")
plt.show()
```
