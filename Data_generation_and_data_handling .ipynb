## Data Generation and Data Handling:

### > Generating Data:

```python
import pandas as pd
import seaborn  as sns
import matplotlib.pyplot as plt
bar_data = pd.read_excel('G:\python_ka_chilla\data_generation_and_handling\Chilla_data2_for_plots_excel.xlsx')
```

### > Handling Data:

#### Graph no # 1
**Given graph is catplot that shows realtion between Gender of people and thier location where they live using generated data:**

```python
sns.set_theme(style="ticks",color_codes=True)
sns.catplot(x = "Gender",y = "Location",hue="Age" ,data=bar_data)
plt.show()
```

#### Graph no # 2
**Given graph is catplot that shows the Age of people and thier qualifications using generated  data:**

```python
sns.set_theme(style="ticks",color_codes=True)
sns.catplot(x = "Age",y = "Qualification_completed" ,data=bar_data)
```

#### Graph no # 3
**Given graph is countplot that shows the Field of people and thier Favorite programming language using generated data:**

```python
sns.set(rc={'figure.figsize':(15,10)})
sns.set_theme(style="ticks",color_codes=True)
p = sns.countplot(x="field_of_study",hue = "Your favorite programming language?",data = bar_data)
plt.legend(loc='upper right',title = "Languages;")
```

#### Graph no # 4
**Given graph is scatter plot that shows the Gender ,Blood group ,Age and Location of people using generated  data:**

```python
sns.set_theme(style="ticks",color_codes=True)
g = sns.FacetGrid(bar_data, row = "Gender",hue="Blood group ")
g = (g.map(plt.scatter,"Age","Location",s =40).add_legend())
```
