```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
sns.set_theme(style="whitegrid")

a= pd.read_excel('G:\python_ka_chilla\data_generation_and_handling\Chilla_data2_for_plots_excel.xlsx')
plt.figure(figsize=(40,20))
g = sns.PairGrid(a, y_vars="Research/Working experience (Float/Int) years",
                 x_vars=["Which mobile sim do you use", "Gender", "Where do you live?", "Which PC are you using?"],
                 height=6, aspect=0.89)

# Draw a seaborn pointplot onto each Axes
g.map(sns.pointplot, scale=0.5, errwidth=3,)
g.set(ylim=(4, 3))
sns.despine(fig=g.fig, left=True)
```
